## Set theory

Set theory deals with the study of sets, which are collections of objects. A set is defined capital letters and as a collection of distinct and well-defined objects, which are called elements or members of the set. Sets can be described using various notations, including list notation, set-builder notation, and interval notation. For example, a set of even numbers can be described as {2, 4, 6, 8, 10, ...}, where the ellipsis (3 dots) represents an infinite sequence of even numbers (Introduction to Sets, n.d.). Set theory is also used in probability theory and statistics, which are essential tools for data analysis (Pishro-Nik, 2014). In these fields, sets are used to represent events and outcomes, and set theory operations are used to calculate probabilities and statistics.
In set theory, a subset is a set that contains only elements that are also in another set, called the superset. In other words, if every element of set B is also an element of set A, then we say that B is a subset of A. The symbol used to denote subset is ⊆. We can also use the phrase "is a subset of" or "is contained in" to describe this relationship.  

In Python, we can use the issubset() method to check if a set is a subset of another set. For example:


```python
# Defining sets A and B
A = {2, 4, 6, 8, 10}
B = {2, 4, 6}

# Checking if A is a subset of B
print(F"Is A subset of B: {A.issubset(B)}") 

# Checking if B is a subset of A
print(F"Is B subset of A: {B.issubset(A)}") 
```

## Set operations
Set operations is a concept similar to fundamental operations on numbers. There are four fundamental operations that can be performed on sets, including union, intersection, complement and difference.
1. Set Union: The union of two sets A and B is the set of all elements that are in A or in B or in both. It is represented by the symbol ∪. In Python, we can use the union() method or the pipe (|) operator to find the union of two sets.


```python
# Set A and B
A = {2, 4, 6, 8, 10}
B = {2, 4, 6}

# Finding the union of sets A and B
C = A.union(B)
print(C) 
# Using pipe (|) operator to find the union of sets A and B
C = A | B
print(C) 
```

2. Set Intersection: The intersection of two sets A and B is the set of all elements that are in both A and B. It is represented by the symbol ∩. In Python, we can use the intersection() method or the ampersand (&) operator to find the intersection of two sets.

```python
# Set A and B form above code

# Finding the intersection of sets A and B
C = A.intersection(B)
print(C) 

# Using ampersand (&) operator to find the intersection of sets A and B
C = A & B
print(C) 
```

3. Set Difference: The difference of two sets A and B is the set of all elements that are in A but not in B. It is represented by the symbol -. In Python, we can use the difference() method or the minus (-) operator to find the difference of two sets.

```python
# Set A and B form above code

# Finding the difference of sets A and B
C = A.difference(B)
print(C)

# Using minus (-) operator to find the difference of sets A and B
C = A - B
print(C) 
```

## Venn diagrams
Venn diagrams are a type of visual representation that is commonly used to show relationships between sets. In a Venn diagram, each set is represented by a circle or another shape, and the overlap between the circles shows the relationship between the sets. For example, if we have two sets, A and B, then the overlap between the circles for A and B represents the elements that are in both A and B. If we have three sets, A, B, and C, then the overlap between the circles for A, B, and C represents the elements that are in all three sets. Venn diagrams can be used to visualize set operations, such as union, intersection, and complement.   

To draw Venn diagrams of the four fundamental set operations in Python, we can use the venn module from the matplotlib_venn library. This library provides a simple way to draw Venn diagrams with customizable labels and colors.


```python
# !pip install matplotlib-venn
from matplotlib_venn import venn2
from matplotlib_venn import venn2,venn3, venn2_circles
```

```python
fig, ax= plt.subplots(2,2, figsize =(12,10))

# Define the sets
set1 = set([1, 2, 3])
set2 = set([2, 3, 4])
set3 = set([1, 4, 6])

# Draw the Venn diagram for Union of two sets
venn2([set1, set2], ax=ax[0,0], set_colors=('blue', 'green'), set_labels=('Set 1', 'Set 2'))
ax[0,0].set_title("Venn diagram")

# Draw the Venn diagram for Intersection of two sets
venn2([set1, set2], set_colors=('blue', 'green'), ax=ax[0,1], set_labels=('Set 1', 'Set 2'))
venn2_circles(subsets=[set1, set2], ax=ax[0, 1])
ax[0,1].set_title('Venn diagram with an outline')

# Draw the Venn diagram for Intersection of two sets
venn2([set1, set2], set_colors=('blue', 'green'), ax=ax[1,0], set_labels=('Set 1', 'Set 2'))
venn2_circles(subsets=[set1, set2], ax=ax[1, 0], linestyle='dashed', linewidth=2, color='k')
ax[1,0].set_title('Venn diagram with a dashed outline')


venn3([set1, set2,set3], set_colors=('blue', 'green',"red"), ax=ax[1,1], set_labels=('Set 1', 'Set 2','Set 3'))
ax[1,1].set_title('Venn diagram with 3 sets')

fig.suptitle("Visualizing and customize your Venn diagram", fontsize=20, y=0.98) 
plt.tight_layout()
plt.savefig('venn.png', dpi=300)
```
